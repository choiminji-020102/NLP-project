```python
### 필요한 라이브러리 임폴트
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
### nltk --> 추가 설치 사항, 필요한 기능을 추가로 다운로드

# 문장을 구분하기 위한 문장 부호 다운로드
nltk.download('punkt')

# 불용어(영어) 다운로드
nltk.download('stopwords')
```

```python
### 문장 단위 토큰화

# text 데이터 생성
text_en = 'God is Great! I won a lottery.'
text_ko = '최근 고지방, 고탄수화물 식이를 즐기는 사람들이 많아지면서 담석증을 호소하는 경우가 늘고 있다.'

# 문장 단위 토큰화 진행 --> sent_tokenize() 사용
# 문장 단위로 잘라서 리스트에 담아줌
en_list = sent_tokenize(text_en)
ko_list = sent_tokenize(text_ko)

# 결과 확인하기
print(f'영어 텍스트 토큰화 결과 확인하기 : {en_list}')  # 2문장
print('-'*80)
print(f'한글 텍스트 토큰화 결과 확인하기 : {ko_list}')  # 1문장
```

```python
### 단어 단위 토큰화

# text 데이터 생성
text_en = 'Learning Python is very exciting and fun.'
text_ko = '오늘은 즐겁고 행복한 수요일 입니다.'

# 단어 단위 토큰화 진행 --> word_tokenize() 사용
# text > 문장 > 단어
en_list = word_tokenize(text_en)
ko_list = word_tokenize(text_ko)

#python --> 문자열 처리 함수 --> split(기준: 기본값=공백): 띄어쓰기(공백) 단위로 분할
'''
이 기준()안에 아무것도 쓰지 않으면 기본값은 공백 -> 띄어쓰기(공백) 단위로 쪼갠다 = 어절 단위로 쪼갠다 => word_tokenize와 비슷
.split()은 문자열 전용 함수임 -> 문자열 뒤에만 쓸 수 있음(문자열.split(기준))
list에 담아서 전달
<word_tokenize와 .split() 유사하지만 다른 점>
- .split(): 앞 단어에 붙어서 처리
- word_tokenize: 문장부호가 별도의 단위로 인식
=> word_tokenize를 굳이 쓰는 이유임. 문장부호를 따로 인식하는 경우가 더 편리함
'''
splited_en = text_en.split()
splited_ko = text_ko.split()

# 결과 확인하기
print(f'영어 텍스트에 대한 단어 토큰화 결과 확인하기 : {en_list}')
print('-'*80)
print(f'영어 텍스트에 대한 split 함수를 적용한 결과 확인하기 : {splited_en}')
print('-'*80)
print(f'한글 텍스트에 대한 단어 토큰화 결과 확인하기 : {ko_list}')
print('-'*80)
print(f'한글 텍스트에 대한 split 함수를 적용한 결과 확인하기 : {splited_ko}')


# 마지막 문장 부호가 별도의 단어로 인식되고 있다
# punkt은 문장 단위 토큰화뿐만 아니라 단어 단위 토큰화에도 사용됨
```

## 한글 텍스트 '형태소' 단위 토큰화 실습



```python
### konlpy 라이브러리 설치
!curl -s https://raw.githubusercontent.com/teddylee777/machine-learning/master/99-Misc/01-Colab/mecab-colab.sh | bash
```

```python
### 필요한 라이브러리(함수) 임폴트
from konlpy.tag import Okt
from mecab import MeCab as Mecab  # c가 대문자임

# 형태소 분석기(함수) 호출
# 필요할 때마다 꺼내쓰기 위해 클래스로 정의되어 있는 함수를 별도의 변수에 반드시 저장함
okt = Okt()
mecab = Mecab()


# nouns: 명사만 뽑아줌
# morphs: 형태소로 쪼갬 (형태소: morpheme)
# pos: 형태소와 품사를 같이 분석
```

```python
### Okt 형태소 분석기 실습

# 실습 문장 생성
text = '한글 형태소 분석기로 테스트를 해보았습니다. 설치 및 동작이 잘 됩니다.'

# 문장 ---> 형태소 단위로 분할(토큰화) ---> morphs() 사용
morphs_okt = okt.morphs(text)

# 문장 ---> 단어 단위로 분할(토큰화) ---> word_tokenize() 사용
tokenized_text = word_tokenize(text)

# 문장 ---> 공백 단위로 분할 ---> split() 사용
# text데이터에 대한 종속 함수, text 데이터가 있어야 사용가능
splited_text = text.split()

# 결과 확인하기
print(f'Okt 형태소 분석기를 이용하여 형태소 분석을 한 결과 : \n{morphs_okt}')
print('-'*80)
print(f'word_tokenize() 함수를 이용하여 단어 단위로 토큰화를 한 결과 : \n{tokenized_text}')
print('-'*80)
print(f'split() 함수를 이용하여 공백 단위로 분할한 결과 : \n{splited_text}')
```

```python
### Mecab 형태소 분석기 실습
# 같은 형태소 분석기이지만 Mecab과 Okt 프로그램 비교

# 실습 문장 생성
text = '한글 형태소 분석기로 테스트를 해보았습니다. 설치 및 동작이 잘 됩니다.'

# 문장 ---> 형태소 단위로 분할(토큰화) ---> morphs() 사용
# Mecab
morphs_mecab = mecab.morphs(text)
# Okt
morphs_okt = okt.morphs(text)

# 결과 확인하기
print(f'Mecab 형태소 분석기를 이용하여 형태소 분석을 한 결과 : \n{morphs_mecab}')
print('-'*80)
print(f'Okt 형태소 분석기를 이용하여 형태소 분석을 한 결과 : \n{morphs_okt}')


# Okt에 비해 Mecab이 빠르기 때문에 더 많이 사용
# 형태소 정의 본질에 더 적합하고 엄밀하게 분할하는 경우가 많음
```

```python
### Okt, Mecab 형태소 분석기 --> 형태소와 품사를 동시에 분석 --> 품사 tagging --> pos() 함수 --> [(형태소, 품사), ...]

# 실습 문장 생성
text = '한글 형태소 분석기로 테스트를 해보았습니다. 설치 및 동작이 잘 됩니다.'

# Okt 형태소 분석기를 이용한 품사 tagging
pos_okt = okt.pos(text)

# Mecab 형태소 분석기를 이용한 품사 tagging
pos_mecab = mecab.pos(text)

# 결과 확인하기
print(f'Okt 형태소 분석기로 형태소와 품사를 동시에 분석한 결과 : \n{pos_okt}')
print('-'*80)
print(f'Mecab 형태소 분석기로 형태소와 품사를 동시에 분석한 결과 : \n{pos_mecab}')
```

```python
### 문정에서 명사만 추출 --> nouns() 사용

# Okt 형태소 분석기를 이용한 명사 추출
nouns_okt = okt.nouns(text)

# Mecab 형태소 분석기를 이용한 명사 추출
nouns_mecab = mecab.nouns(text)

# 결과 확인하기
print(f'Okt 형태소 분석기를 이용하여 명사만 추출한 결과 : \n{nouns_okt}')
print('-'*80)
print(f'Mecab 형태소 분석기를 이용하여 명사만 추출한 결과 : \n{nouns_mecab}')
```

```python
### 실습에 필요한 라이브러리 설치

# Hugging Face에서 제공하는 tokenizers 라이브러리 설치
!pip install tokenizers
```

```python
### 필요한 함수 임폴트
from tokenizers import BertWordPieceTokenizer
```

```python
### BertWordPieceTokenizer 클래스 함수 호출
wordpiece_tokenizer = BertWordPieceTokenizer()
```

```python
### BertWordPieceTokenizer 학습하기

# 파일 경로 설정
file_path = '/content/drive/MyDrive/딥러닝/nlp/nsmc.txt'

# BertWordPieceTokenizer 학습 --> model.train()
wordpiece_tokenizer.train(
    files = file_path,
    vocab_size = 10000,
    min_frequency = 1
)
```

```python
### 생성된 단어 사전 확인 --> get_vocab() 함수
# vocab_10000 = wordpiece_tokenizer.get_vocab()
vocab_1000 = wordpiece_tokenizer.get_vocab()

# 단어 사전의 자료형 확인
print(f'단어 사전의 자료형 = {type(vocab_1000)}')

print('-' * 80)

# 단어 사전 출력
print(vocab_1000)
```

```python
### 학습된 단어 사전 저장하기
file_path = '/content/drive/MyDrive/딥러닝/nlp'
wordpiece_tokenizer.save_model(file_path)
```

```python
### WordPiece 토크나이저 적용하기

# 테스트 문장
text = '너와 함께한 시간 모두 눈부셨다. 날이 좋아서 날이 좋지 않아서 날이 적당해서 모든 날이 좋았다.'

# 토크나이저 : 텍스트 --> 토큰화(subword) --> 단어 사전 참조 --> 숫자화(임베딩) => encode() 함수
encoded_text = wordpiece_tokenizer.encode(text)

# 결과 확인하기
print(encoded_text)
```

```python
### 단어 사전의 단어 수 = 10000개의 Encoding 결과 --> 속성(attributes) --> ids, tokens 속성값 확인

# ids 확인
print(f'ids = \n{encoded_text.ids}')
print('-' * 80)

# tokens 확인
print(f'tokens = \n{encoded_text.tokens}')
```

```python
### 단어 사전의 단어 수 = 1000개의 Encoding 결과 --> 속성(attributes) --> ids, tokens 속성값 확인

# ids 확인
print(f'ids = \n{encoded_text.ids}')
print('-' * 80)

# tokens 확인
print(f'tokens = \n{encoded_text.tokens}')
```

```python
### 한글로 복원 --> decode() 함수 적용
decoded_text = wordpiece_tokenizer.decode(encoded_text.ids)

# 결과 확인하기
print(f'한글로 복원한 결과 : \n{decoded_text}')
```
