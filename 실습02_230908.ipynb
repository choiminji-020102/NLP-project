### CounterVectorizer 실습

```python
### 필요한 라이브러리 임폴트
from sklearn.feature_extraction.text import CountVectorizer
```

```python
### 클래스 함수 호출
count = CountVectorizer()
```

```python
### 학습 전 --> 단어 사전 생성 여부 확인 --> vocabulary_
print(count.vocabulary_)

# => 생성X
```

```python
### 한글 실습용 텍스트 데이터 생성
text = ['나는 배가 고프다', '내일 점심 뭐 먹지', '내일 공부 해야겠다', '점심 먹고 공부해야지']
# 여러개의 데이터는 묶어서 저장해야 함. (대괄호나 소괄호)리스트나 튜플의 형태로 담아놔야 함
```

```python
### 단어 사전 생성

# fit() 실행 --> 학습
vocab = count.fit(text)

# 생성된 단어 사전 --> vocabulary_ 변수에 저장
# 생성된 단어 사전 확인 --> count.vocabulary_
vocab = count.vocabulary_  # 변수 안에 변수가 들어있음

# 결과 확인
print(vocab)

# 학습을 통해 단어사전을 만들면 vocabulary_라는 변수로 저장되기 때문에 변수를 통해 꺼내써야함
# pyton3.6 이전 버전이기 때문에 dict의 순서 무질서함
```

```python
### 임베딩을 할 문장 설정

# 1개 문장 설정 --> 인덱싱
sentence = text[0]

# 문장 --> 임베딩
# embedding = count.transform(sentence)
# print(embedding)  => 에러!


# 다시 -----------------------------------


# 1개 문장 설정 --> 인덱싱
sentence_list = [text[0]]
print(sentence_list)

print('-' * 80)

# 문장 --> 임베딩
embedding = count.transform(sentence_list).toarray()
print(embedding)
print('-' * 80)
embedding = count.transform(sentence_list)
print(embedding)
```

### TF-IDF 실습

```python
'''
 1. 기본적인 방법은 CountVecterizer와 유사하다.
 - 단어 사전을 만들어서, 단어 사전에 등장하는 단어를 적용하는 것은 동일하다.

 2. 단어의 출현 횟수가 아니라 각 단어의 TF_IDF값을 적용하는 것이 다르다.
'''

# 필요한 라이브러리 임폴트
from sklearn.feature_extraction.text import TfidfVectorizer

# TfidfVectorizer 함수 호출
tfidf = TfidfVectorizer()
```

```python
### 실습용 텍스트 데이터 생성 및 단어 사전 생성하기

# 한글 실습용 텍스트 데이터 생성
text = ['나는 배가 고프다', '내일 점심 뭐 먹지', '내일 공부 해야겠다', '점심 먹고 공부해야지']

# fit() 함수를 이용 ---> 단어 사전 생성
tfidf.fit(text)

# 생성된 단어 사전 확인하기 (학습된 모델 내부에 저장되어 있음)
vocab = tfidf.vocabulary_
print(vocab)
```

```python
### 문장 임베딩 --> 2차원 구조(행렬)

# 실습용 문장 생성 --> 리스트 자료형 데이터 생성 --> 1차원 데이터 생성 : ["나는 배가 고프다"]
sentence_list = [text[0]]
print(f'실습용 문장 = {sentence_list}')

print('-' * 80)

# transform().toarray() 함수 --> 문장의 행렬화 완성
embedding = tfidf.transform(sentence_list).toarray()
print(f'임베딩의 결과 : \n{embedding}')
```

## 텍스트 유사도 측정 - 코사인 유사도

```python
### 필요한 함수 임폴트
from sklearn.metrics.pairwise import cosine_similarity
```

```python
### TfidfVectorizer 함수 호출
tfidf = TfidfVectorizer()
```

```python
### 실습용 텍스트 데이터 생성
text = ["직원이 무단 퇴사를 했는데 손해 배상 청구할 수 있나요?", "무단 퇴사한 직원에 대한 손해 배상 청구가 가능한가요?"]
```

```python
### 문장 임베딩 : 문장의 행렬화

# fit() --> 단어 사전 생성
tfidf.fit(text)

# 생성된 단어 사전 확인
vocab = tfidf.vocabulary_
print(f'생성된 단어 사전 : \n{vocab}')

print('-'*80)

# 두 문장 --> 각각 1차원 구조의 데이터 변환
sentence1 = [text[0]]
sentence2 = [text[1]] # text[1]는 문자열 그 자체이기 때문에 []를 씌워 리스트로 만듬

print(f'코사인 유사도를 측정할 첫번째 문장 : \n{sentence1}')
print('-'*80)
print(f'코사인 유사도를 측정할 두번째 문장 : \n{sentence2}')

print('-'*80)

# transform().toarray() 사용
embedding1 = tfidf.transform(sentence1).toarray()
embedding2 = tfidf.transform(sentence2).toarray()
print(f'첫번째 문장의 임베딩 행렬 : \n{embedding1}')
print('-'*80)
print(f'두번째 문장의 임베딩 행렬 : \n{embedding2}')
```

```python
### (두 문장 간의) 코사인 유사도 측정

# cosine_similarity(X, Y) 사용
X = embedding1
Y = embedding2
similarity = cosine_similarity(X, Y)

# 결과 확인하기
print(similarity)
```

```python
### 학습 데이터 생성하기

# 파일 경로 설정
file_path = '/content/drive/MyDrive/Colab Notebooks/딥러닝-이영배T/데이터/nsmc.txt'

# 학습 데이터
train_df = pd.read_table(file_path)

# 결과 확인
train_df  # print()했을 때보다 가독성이 좋음
```

```python
### df.col.str.replace(pattern,repl) --> 정규 표현식의 기능 구현 --> 한글 글자(가-힣)외 문자 제거
train_df1.document = train_df1.document.str.replace('[^가-힣]', ' ')

# 결과 확인하기
train_df1
```

### Okt 형태소 분석기를 사용한 word2vec 모델 학습

```python
### Okt 형태소 분석기를 사용한 토큰화 작업 + 불용어 제거

# 불용어 정의
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']

# okt.morphs() 함수를 이용한 토큰화 + 불용어 제거
# => [[리뷰1]...[리뷰100000]]
# => [리뷰1] = [토큰1, 토큰2...토큰n]
tokenized_data1 = []  # 전체 결과 담을 리스트
for sentence in train_df2.document:
  sentence_tokenized = []  # 리뷰 하나에 대해 토큰화한 후 불용어 제거한 토큰 리스트
  # 토큰화
  tokenized = okt.morphs(sentence)  # 리뷰 하나에 대해 불용어가 포함되어 있는 전체 토큰 리스트
  # 불용어 제거
  for token in tokenized:
    if token not in stopwords:
      sentence_tokenized.append(token)
  tokenized_data1.append(sentence_tokenized)

# tokenized_data1 => list of lists of tokens
```

```python
### 불용어를 제거하고 토큰화를 해준 데이터 확인 ---> 상위 5개(문장에 대한 토큰들) 확인
tokenized_data1[:5]
```

```python
### CBOW, skip-gram 모델 학습 --> Word2Vec(매개변수=값) --> 토큰별 임베딩 벡터 생성

# 저장 경로 설정
cbow_save_path = '/content/drive/MyDrive/딥러닝/nlp/cbow_okt.model'
sg_save_path = '/content/drive/MyDrive/딥러닝/nlp/sg_okt.model'

# CBOW 모델 학습
model_cbow_okt = Word2Vec(sentences=tokenized_data1, vector_size=100, window=5, min_count=5, workers=1, sg=0, seed=0)

# skip-gram 모델 학습
model_sg_okt = Word2Vec(sentences=tokenized_data1, vector_size=100, window=5, min_count=5, workers=1, sg=1, seed=0)

# CBOW 모델 저장
model_cbow_okt.save(cbow_save_path)

# skip-gram 모델 저장
model_sg_okt.save(sg_save_path)
```

```python
### CBOW, skip-gram 모델 불러오기 --> load() 함수 사용

# 저장 경로 설정
cbow_save_path = '/content/drive/MyDrive/딥러닝/nlp/cbow_okt.model'
sg_save_path = '/content/drive/MyDrive/딥러닝/nlp/sg_okt.model'

# CBOW 모델 불러오기
loaded_cbow_okt = Word2Vec.load(cbow_save_path)

# skip-gram 모델 불러오기
loaded_sg_okt = Word2Vec.load(sg_save_path)
```

```python
### CBOW 모델 --> 생성된 단어 사전과 임베딩 벡터(wv) 확인하기

'''
1. 임베딩 벡터 생성 --> model.wv
2. 단어 사전 : model.wv.key_to_index
3. model.wv['word']를 통해 특정 단어의 embedding vector를 확인할 수 있다
4. model.wv.vectors 속성을 이용하여 생성된 단어 사전의 전체 value(embedding vector)
를 확인할 수 있다.
5. model.wv.similarity(w1=w1, w2=w2) 메서드를 이용하여 단어 간의 유사도를 비교할 수 있다.
6. model.wv.most_similar(word, topn=n) 메서드를 이용하여 특정 단어와 가장 가까운 단어
n개를 추출할 수 있다.
'''

# 학습된 모델 --> 생성물 wv 확인
print(loaded_cbow_okt.wv)

print('-'*80)

# 단어 사전 확인하기 --> model.wv.key_to_index
vocab = loaded_cbow_okt.wv.key_to_index
print(f'단어 사전의 자료형 = {type(vocab)}')
print('-'*80)
print(f'단어 사전 : \n {vocab}')

print('-'*80)

# 단어 사전 --> 전체 key(token) 추출 --> model.wv.key_to_index.keys() 함수
tokens = vocab.keys()
print(f'전체 token 확인 : \n{tokens}')

print('-'*80)

# 특정 token --> 임베딩 벡터 추출
embedding = loaded_cbow_okt.wv['안개']
print(f'안개라는 token의 임베딩 벡터 : \n{embedding}')

print('-'*80)

# token 전체 --> 임베딩 벡터 추출
embeddings = loaded_cbow_okt.wv.vectors
print(f'token 전체에 대한 임베딩 벡터의 자료형 : {type(embeddings)}')
print('-'*80)
print(f'token 전체에 대한 임베딩 벡터 모양 확인 : \n{embeddings.shape}')
```

### Mecab 형태소 분석기를 사용한 word2vec 모델 학습

```python
### Mecab 형태소 분석기를 사용한 토큰화 작업 + 불용어 제거

# 불용어 정의
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']

# okt.morphs() 함수를 이용한 토큰화 + 불용어 제거
tokenized_data2 = []
for sentence in train_df2.document:
    sentence_tokenized = []
    # 토큰화
    tokenized = mecab.morphs(sentence)
    # 불용어 제거
    for token in tokenized:
        if token not in stopwords:
            sentence_tokenized.append(token)
    tokenized_data2.append(sentence_tokenized)
```

```python
### 불용어를 제거하고 토큰화를 해준 데이터 확인 --> 상위 5개 확인
tokenized_data2[:5]
```

```python
### CBOW, skip-gram 모델 학습 --> Word2Vec(매개변수=값) --> 토큰별 임베딩 벡터 생성

# 저장 경로 설정
cbow_save_path = '/content/drive/MyDrive/딥러닝/nlp/cbow_mecab.model'
sg_save_path = '/content/drive/MyDrive/딥러닝/nlp/sg_mecab.model'

# cbow 모델 학습
model_cbow_mecab = Word2Vec(sentences=tokenized_data2, vector_size=100, window=5, min_count=5,workers=1, sg=0, seed=0)

# skip-gram 모델 학습
model_sg_mecab = Word2Vec(sentences=tokenized_data2, vector_size=100, window=5, min_count=5,workers=1, sg=1, seed=0)

# CBOW 모델 저장
model_cbow_mecab.save(cbow_save_path)

# skip-gram 모델 저장
model_sg_mecab.save(sg_save_path)
```

```python
### CBOW, skip-gram 모델 불러오기 --> load() 함수 사용

# 저장 경로 설정
cbow_save_path = '/content/drive/MyDrive/딥러닝/nlp/cbow_mecab.model'
sg_save_path = '/content/drive/MyDrive/딥러닝/nlp/sg_mecab.model'

# CBOW 모델 불러오기
loaded_cbow_mecab = Word2Vec.load(cbow_save_path)

# skip-gram 모델 불러오기
loaded_sg_mecab = Word2Vec.load(sg_save_path)
```

```python
### CBOW 모델 --> 생성된 단어 사전과 임베딩 벡터(wv) 확인하기

'''
1. 임베딩 벡터 생성 --> model.wv
2. 단어 사전 : model.wv.key_to_index
3. model.wv['word']를 통해 특정 단어의 embedding vector를 확인할 수 있다
4. model.wv.vectors 속성을 이용하여 생성된 단어 사전의 전체 value(embedding vector)
를 확인할 수 있다.
5. model.wv.similarity(w1=w1, w2=w2) 메서드를 이용하여 단어 간의 유사도를 비교할 수 있다.
6. model.wv.most_similar(word, topn=n) 메서드를 이용하여 특정 단어와 가장 가까운 단어
n개를 추출할 수 있다.
'''

# 학습된 모델 --> 생성물 wv 확인
print(loaded_cbow_mecab.wv)

print('-'*80)

# 단어 사전 확인하기 --> model.wv.key_to_index
vocab = loaded_cbow_mecab.wv.key_to_index
print(f'단어 사전의 자료형 = {type(vocab)}')
print('-'*80)
print(f'단어 사전 : \n {vocab}')

print('-'*80)

# 단어 사전 --> 전체 key(token) 추출 --> model.wv.key_to_index.keys() 함수
tokens = vocab.keys()
print(f'전체 token 확인 : \n{tokens}')

print('-'*80)

# 특정 token --> 임베딩 벡터 추출
embedding = loaded_cbow_mecab.wv['안개']
print(f'안개라는 token의 임베딩 벡터 : \n{embedding}')

print('-'*80)

# token 전체 --> 임베딩 벡터 추출
embeddings = loaded_cbow_mecab.wv.vectors
print(f'token 전체에 대한 임베딩 벡터의 자료형 : {type(embeddings)}')
print('-'*80)
print(f'token 전체에 대한 임베딩 벡터 모양 확인 : \n{embeddings.shape}')
```
