```python
### 문장 임베딩 벡터를 생성하기 위해서 한글로 파인 튜닝된 SentenceBERT 모델 생성하기
model_name = 'ddobokki/klue-roberta-base-nli-sts'
model = SentenceTransformer(model_name)  # SentenceBERT model
```

```python
### 모델 테스트

# 테스트 문장 생성
sentence = "본 발명은 바위수염 추출물을 포함하는 항비만용 조성물 및 그 제조방법에 관한 것이다"

# encode() 이용 --> 문장 임베딩 벡터 생성
embedding = model.encode(sentence)

# 결과 확인하기
print(f'임베딩 벡터의 모양 : {embedding.shape}')
print('-'*80)
print(f'임베딩 벡터 확인 : \n{embedding}')
```

```python
### 데이터 -- 임베딩 벡터 생성 및 저장하기

'''
1. apply() 함수, lambda 함수를 동시에 사용
2. 문장 임베딩 벡터를 생성하고 numpy ndarray 자료형으로 저장하기
'''

# embeddings = []

# for sentence in df.Q:
#   print(sentence)
#   embedding = model.encode(sentence)
#   embeddings.append(embedding)

# 문장 임베딩 벡터를 생성하고 numpy ndarray 자료형으로 변수에 저장하기
embeddings = df.Q.apply(lambda x : model.encode(x)).values
# Series에 문장이 있던 자리에 임베딩 벡터가 채워짐 -> 자료형: Series
# .values -> 자료형: (넘파이)배열

# 결과 확인하기
print(f'문장 임베딩 벡터 전체의 모양 : {embeddings.shape}')
print('-'*80)
print(f'문장 임베딩 벡터 1개의 자료형 : {type(embeddings[0])}, \n 문장 임베딩 벡터 1개의 모양 : {embeddings[0][0].shape}')

# 결과 저장하기
embedding_path = '/content/drive/MyDrive/Colab Notebooks/딥러닝-이영배T/데이터/chatbot_embeddings.npy'
np.save(embedding_path, embeddings)
```

```python
### 두 문장의 텍스트 유사도 계산 함수 정의

# 필요한 라이브러리 임폴트
from numpy import dot
from numpy.linalg import norm

def cos_sim(A, B):
    return dot(A, B) / (norm(A)*norm(B))
```

```python
### 사용자 질문에 대한 응답 구현

'''
1. 사용자의 질문이 들어오면 해당 질문에 대한 문장 임베딩 벡터 값을 구한다.
2. 사용자 질문의 문장 임베딩 벡터 값과 챗봇 데이터의 임베딩 컬럼(train_data['embedding'])에
저장해둔 모든 질문 샘플들의 문장 임베딩 벡터 값들을 이용, 코사인 유사도를 계산
--> 코사인 유사도 값을 저장하는 컬럼(train_data['score'])을 생성하고 값을 저장
3. 코사인 유사도 값이 가장 높은 질문 샘플을 찾는다.
4. 해당 질문 샘플과 짝이 되는 답변 샘플을 반환한다.
'''

def chatbot(question):
    embedding = model.encode(question)
    # 코사인 유사도 계산 --> 'score'이름의 컬럼 생성 --> 유사도 값 저장
    df['score'] = df.embedding.apply(lambda x: cos_sim(x, embedding))
    # score 컬럼 값(코사인 유사도 값)이 가장 큰 행 추출 --> A 컬럼의 값 --> 답변으로 선택
    row = df.score.idxmax()
    # print(f'사용자가 입력한 문장과 코사인 유사도 값이 가장 큰 행의 인덱스 : {row}')
    answer = df.loc[row, 'A']
    return answer
```

```python
### 챗봇 성능 테스트(1)
chatbot('반가워')
```
