## SentenceBERT를 이용한 문장 유사도 측정

```python
### 필요한 라이브러리 설치 : sentence_transformers (하드디스크에 설치)
!pip install sentence_transformers
```

```python
### 필요한 라이브러리 임폴트 (RAM에 올려둠)
from sentence_transformers import SentenceTransformer
```

### 한글 문장 유사도 측정

```python
### 문장 임베딩 벡터 추출 --> 파인 튜닝한 한국어 SentenceBERT(SentenceTransformer) 모델 생성
model_name = 'ddobokki/klue-roberta-base-nli-sts'
model_kor = SentenceTransformer(model_name)
```

```python
### 한글 텍스트 데이터 생성
ko_sentences = ['직원이 무단 퇴사를 했는데 손해 배상 청구할 수 있나요?', '무단 퇴사한 직원에 대한 손해 배상 청구가 가능한가요?']
```

```python
### 문장 임베딩 벡터 생성하기
embeddings = model_kor.encode(ko_sentences)

# 결과 확인하기
print(f'실습용 문장의 임베딩 행렬의 모양 : {embeddings.shape}')
print('-'*80)
print(f'실습용 문장 1의 임베딩 벡터 : \n{embeddings[0]}')
print('-'*80)
print(f'실습용 문장 2의 임베딩 벡터 : \n{embeddings[1]}')
```

```python
### 코사인 유사도 측정

# 필요한 함수 임폴트
from sklearn.metrics.pairwise import cosine_similarity

# 1차원 배열 --> 2차원 행렬로 변환
embedding1 = embeddings[0].reshape((1,768))
embedding2 = embeddings[1].reshape((1,768))

# cosine_similarity 함수 사용
sim = cosine_similarity(embedding1, embedding2)

# 결과 확인하기
print(sim)
```

```python
### pd.read_csv() --> DataFrame 자료형 데이터 생성

# 필요한 라이브러리 임폴트
import pandas as pd

# 파일 경로 설정
file_path = '/content/drive/MyDrive/Colab Notebooks/딥러닝/데이터/movies_metadata.csv'

# DataFrame 자료형 생성
df_movies = pd.read_csv(file_path)

# 결과 확인하기
print(df_movies)
```

```python
### 기본 정보 확인 --> df.info()
df_movies.info()

# overview : 44512개
# title : 45460개
# => 누락 데이터 제거 필요!
```

```python
### 줄거리, 제목 컬럼 --> 팬시 인덱싱
data = df_movies.loc[:, ['overview', 'title']]

# 결과 확인하기
print(data)
```

```python
### 누락 '행' 제거 --> dropna(axis=0)
df_cleaned = data.dropna()

# 결과 확인하기
print(df_cleaned)
print('-'*80)
print(f'각 컬럼 별 누락 데이터의 수 : \n{df_cleaned.isnull().sum()}')
```

```python
### 누락 데이터 제거 후 인덱스 재조정 --> df.reset_index(drop=True, inplace=True)
df_cleaned.reset_index(drop=True, inplace=True)

# 결과 확인하기
print(df_cleaned)
```

```python
### data의 수를 20000개로 축소
df_data = df_cleaned.iloc[0:20000, :]

# 결과 확인하기
print(df_data)
```

```python
### SentenceBERT 모델 생성
model_name = 'sentence-transformers/all-MiniLM-L12-v2'
model = SentenceTransformer(model_name)
```

```python
### 문장 임베딩 벡터 생성

# overview 컬럼 --> df_data.overview.values
sentences = df_data.overview.values  # df_data['overview'].values와 같음
print(f'overview 컬럼에서 추출한 값 확인 : \n{sentences}')

print('-'*80)

# model.encode() 함수 사용
embeddings = model.encode(sentences)

# 결과 확인하기
print(f'생성된 문장 임베딩 행렬의 모양 : {embeddings.shape}')
print('-'*80)
print(f'문장 1개의 임베딩 벡터의 모양 : {embeddings[0].shape}')
```

```python
### 문장 임베딩 행렬 저장

# 필요한 라이브러리 임폴트
import numpy as np

# 데이터 저장
np.save('/content/drive/MyDrive/Colab Notebooks/딥러닝/nlp/embeddings.npy', embeddings)
```

```python
### 저장된 넘파이 배열(임베딩 행렬) 불러오기

# 필요한 라이브러리 임폴트
import numpy as np
import pandas as pd

# 파일 경로 설정
file_path = '/content/drive/MyDrive/Colab Notebooks/딥러닝/nlp/embeddings.npy'

# 저장된 넘파이 배열 불러오기
embeddings = np.load(file_path)

# 결과 확인하기
print(f'문장 임베딩 행렬의 모양 : {embeddings.shape}')
```

```python
### linear_kernel() 함수 --> 20000개의 영화 줄거리 각각에 대해서 코사인 유사도 측정
# 코사인 유사도를 대신함. 집단적으로 계산할 때 사용함.

# 필요한 함수 임폴트
from sklearn.metrics.pairwise import linear_kernel

# 20000개의 영화 줄거리간의 코사인 유사도 계산 --> 유사도 행렬(자료형 : 데이터프레임) 생성
sim_matrix = linear_kernel(embeddings, embeddings)

# 결과 확인하기
print(f'유사도 측정의 결과 : \n{sim_matrix}')
print('-'*80)
print(f'유사도 측정의 결과 값의 모양 : {sim_matrix.shape}')
print('-'*80)
print(f'Toy Story에 대한 유사도 측정의 결과 : \n{sim_matrix[:, 0]}')
```

```python
### 코사인 유사도 행렬 --> 데이터프레임 생성

# 데이터프레임 생성
columns = df_data.title.values
index = df_data.title.values

df_sim = pd.DataFrame(data=sim_matrix, columns=columns, index=index)

# 결과 확인하기
df_sim
```

```python
### 특정 영화 기준 --> 줄거리가 유사한 영화 추천 함수 생성

def recommend(title):
  top10 = df_sim.loc[:, title].sort_values(ascending=False).iloc[1:11]
  return top10

# 결과 확인하기
top10 = recommend('GoldenEye')
print(f'결과 확인하기 : \n{top10}')
```
